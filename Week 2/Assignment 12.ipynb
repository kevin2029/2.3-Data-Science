```python
import scipy.stats as st
import pandas as pd
import seaborn as sns
videoGamesSales = pd.read_csv('Datasets/vgsales.csv', sep= ',')
```

```python
videoGamesSales.head()
```

```python
videoGamesSales.groupby('Genre').mean()
```

```python
graph = sns.barplot(y='Global_Sales', x='Genre', data=videoGamesSales)
graph.set_xticklabels(graph.get_xticklabels(), rotation=90)
```

```python
graph = sns.barplot(y='NA_Sales', x='Year', data=videoGamesSales)
graph.set_xticklabels(graph.get_xticklabels(), rotation=90)
```

```python
graph = sns.barplot(y='Global_Sales', x='Platform', data=videoGamesSales)
graph.set_xticklabels(graph.get_xticklabels(), rotation=90)
```
